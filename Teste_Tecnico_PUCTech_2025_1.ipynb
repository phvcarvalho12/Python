````python

# Seu c√≥digo come√ßa aqui:
# Instalar bibliotecas
!pip install seaborn --quiet

# Importa√ß√µes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Configura√ß√µes
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)


# 1. Carregar dados
url = "https://raw.githubusercontent.com/puc-tech/challenge/refs/heads/main/student_preferences_extended.csv"
df = pd.read_csv(url)

print("üìå Formato original:", df.shape)
display(df.head())

print("\nüìå Tipos de dados:")
print(df.dtypes)

print("\nüìå Valores ausentes por coluna:")
print(df.isnull().sum())

# 2. Limpeza dos dados

## 2.1 Remover colunas com mais de 50% de valores nulos
limiar_nulos = 0.5
colunas_com_muitos_nulos = df.columns[df.isnull().mean() > limiar_nulos]
df.drop(columns=colunas_com_muitos_nulos, inplace=True)
print(f"\n‚ùå Colunas removidas por muitos valores nulos (>50%): {list(colunas_com_muitos_nulos)}")

## 2.2 Remover colunas com baixa variabilidade (1 valor √∫nico)
colunas_baixa_variabilidade = [col for col in df.columns if df[col].nunique() <= 1]
df.drop(columns=colunas_baixa_variabilidade, inplace=True)
print(f"\n‚ùå Colunas removidas por baixa variabilidade (1 valor √∫nico): {colunas_baixa_variabilidade}")

## 2.3 Remover linhas duplicadas
linhas_duplicadas = df.duplicated().sum()
df.drop_duplicates(inplace=True)
print(f"\n‚ùå Linhas duplicadas removidas: {linhas_duplicadas}")

print("\n‚úÖ Formato ap√≥s limpeza:", df.shape)

# 3. An√°lises
print("\nüìä Estat√≠sticas descritivas:")
display(df.describe(include='all'))


# 3.1. Colunas categ√≥ricas
categoricas = [col for col in df.select_dtypes(include='object').columns
               if df[col].nunique() <= 50 and col != 'id_aluno']

print("\nüìå Colunas categ√≥ricas para gr√°fico de barras:", categoricas)

for col in categoricas:
    plt.figure()
    df[col].value_counts(dropna=False).plot(kind='bar')
    plt.title(f'Distribui√ß√£o: {col}')
    plt.ylabel('Frequ√™ncia')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 3.2. Colunas num√©ricas
numericas = df.select_dtypes(include=np.number).columns
print("\nüìå Colunas num√©ricas:", list(numericas))

for col in numericas:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribui√ß√£o: {col}')
    plt.tight_layout()
    plt.show()

# 3.3. Correla√ß√£o
if len(numericas) > 1:
    plt.figure(figsize=(10, 8))
    sns.heatmap(df[numericas].corr(), annot=True, cmap='coolwarm')
    plt.title("Matriz de Correla√ß√£o")
    plt.show()

# 3.4. Boxplots para outliers
for col in numericas:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot - {col}')
    plt.tight_layout()
    plt.show()

# 3.5. Colunas restantes com muitos nulos (alerta leve)
restantes_nulos = df.columns[df.isnull().sum() > 0]
if len(restantes_nulos) > 0:
    print(f"\n‚ö†Ô∏è Colunas com valores nulos restantes (mas <50%): {list(restantes_nulos)}")


# 4. Visualiza√ß√µes
#print("\nüßæ Visualiza√ß√£o final dos dados (amostra):")
#display(df.head(10))


# 5. Relat√≥rio final
print("\n‚úÖ Conclus√£o da an√°lise:")
print("- Colunas in√∫teis e linhas duplicadas foram removidas.")
print("- Tabelas limpas e prontas para an√°lise ou modelagem.")
print("- Visualiza√ß√µes geradas para compreens√£o dos dados.")
````
